```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
ras_18_normal = np.array([37.41, 40.14, 39.13, 33.97, 16.81])
ras_18_signed = np.array([37.48, 39.42, 39.59, 35.66, 17.3])
```

```python
x = np.array([10, 30, 50, 70, 90])
```

```python
ras_34_normal = np.array([40.22, 39.44, 38.61, 35.08, 19.92])
ras_34_signed = np.array([38.27, 40.1, 39.69, 34.15, 21.1])
```

```python
unet_normal = np.array([89.152, 90.1091, 89.912, 89.6415, 88.738])
unet_normal_bilinear = np.array([89.6713, 90.5642, 90.4589, 90.0764, 90.1028])
unet_signed = np.array([87.2491, 90.2405, 90.2103, 90.0805, 89.6934])
unet_signed_bilinear = np.array([89.5345, 90.331, 90.483, 90.3719, 89.5378])
```

```python
vgg11bn_normal = np.array([80.97, 83.7, 83.26, 79.13, 53.97])
vgg11bn_signed = np.array([82.07, 84.23, 82.72, 79.39, 50.51])
```

```python
plt.figure(figsize=(7,6))
plt.title("VGG-11, with Batchnorm", fontsize=21)
plt.plot(x, vgg11bn_normal,'-o', label = "Kaiming Normal")
plt.plot(x, vgg11bn_signed,'-o', label = "Signed Kaiming constant")
plt.grid()
plt.xticks(x, fontsize=21);
plt.yticks(fontsize=21);
plt.axhline(y=87.17, color='r', linestyle='-', linewidth=3, label='Accuracy of the dense model')
plt.legend(fontsize=18);
plt.xlabel("% of Weights", fontsize=25)
plt.ylabel('Accuracy', fontsize=25)
plt.rcParams['axes.facecolor'] = 'white'
plt.savefig('pic3.png')
```

```python
listt=[[ras_18_normal,ras_18_signed,"ResNet-18","Kaiming Normal","Signed Kaiming constant"],
       [ras_34_normal,ras_34_signed,"ResNet-34"],[unet_normal,unet_signed,"U-Net, Transposed Convolution"],
       [unet_normal_bilinear,unet_signed_bilinear,"U-Net, Bilinear Upsample"],[vgg11bn_normal,vgg11bn_signed,"VGG11BN"]]
```

```python
fig, axes = plt.subplots(ncols=2, figsize=(15, 6))
for ls,ax in zip(listt[0:2], axes.flat):
    ax.set_title(ls[2], fontsize=25)
    ax.plot(x, ls[0], '-o', label = listt[0][3])
    ax.plot(x, ls[1], '-o', label = listt[0][4])
    ax.legend();
    ax.grid()
    ax.set_xticks(x);
    ax.set_ylim(20.0,46)
    ax.tick_params(axis = 'both', which = 'major', labelsize = 19)
    ax.set_xlabel("% of Weights", fontsize=25)
    ax.set_ylabel('Accuracy', fontsize=25)
axes.flat[0].axhline(y=44, color='r', linestyle='-', linewidth=3, label='Accuracy of the dense model')
axes.flat[0].legend(fontsize=18)
axes.flat[0].set_facecolor('xkcd:white')
axes.flat[1].axhline(y=45.4, color='r', linestyle='-', linewidth=3, label='Accuracy of the dense model')
axes.flat[1].legend(fontsize=17)
axes.flat[1].set_facecolor('xkcd:white')
plt.savefig('pic1.png')
```

```python
fig, axes = plt.subplots(ncols=2, figsize=(15, 6))
for ls,ax in zip(listt[2:4], axes.flat):
    ax.set_title(ls[2], fontsize=25)
    ax.plot(x, ls[0], '-o', label = listt[0][3])
    ax.plot(x, ls[1], '-o', label = listt[0][4])
    ax.legend();
    ax.grid()
    ax.set_xticks(x);
    ax.set_ylim(88.0,91.6)
    ax.tick_params(axis = 'both', which = 'major', labelsize = 19)
    ax.set_xlabel("% of Weights", fontsize=25)
axes.flat[0].set_ylabel('Accuracy', fontsize=21)
axes.flat[0].axhline(y=91.47, color='r', linestyle='-', linewidth=3, label='IoU of the dense model')
axes.flat[0].legend(fontsize=17)
axes.flat[0].set_facecolor('xkcd:white')
axes.flat[1].axhline(y=91.3, color='r', linestyle='-', linewidth=3, label='IoU of the dense model')
axes.flat[1].legend(fontsize=14)
axes.flat[1].set_facecolor('xkcd:white')
plt.savefig('pic2.png')
```
